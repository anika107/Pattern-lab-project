```
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
```

```
#ENAC

#5289 (training)
positive_5289 = "/content/drive/My Drive/Colab Notebooks/Pattern Recognition/5289_thaliana .npy files/Positive/A.thaliana5289_pos_ENAC_4000.npy"
negative_5289 = "/content/drive/My Drive/Colab Notebooks/Pattern Recognition/5289_thaliana .npy files/Negative/A.thaliana5289_neg_ENAC_4000.npy"

#1000 (test)
positive_1000 = "/content/drive/My Drive/Colab Notebooks/Pattern Recognition/1000_thaliana .npy files/Positive/1000_pos_ENAC.npy"
negative_1000 = "/content/drive/My Drive/Colab Notebooks/Pattern Recognition/1000_thaliana .npy files/Negative/1000_neg_ENAC.npy"

#x_training and x_testing
x_train_positive = np.load(positive_5289)
x_train_negative = np.load(negative_5289)
x_test_positive = np.load(positive_1000)
x_test_negative = np.load(negative_1000)

#y_training and y_testing
y_train_positive = np.tile(1,5289)
y_train_negative = np.tile(0,5289)
y_test_positive = np.tile(1,1000)
y_test_negative = np.tile(0,1000)


#1000 instances, 148 features
#5289 instance, 148 features

#printing the content number of the npy files
print(x_test_negative.shape)
print(y_test_negative.shape)

print(x_train_negative.shape)
print(y_train_negative.shape)
```

```
seed = 40
np.random.seed(seed)
```

```
#training set
x_train_s = np.concatenate((x_train_positive, x_train_negative), axis=0)
y_train_s = np.concatenate((y_train_positive, y_train_negative))

#test set
x_test_s = np.concatenate((x_test_positive, x_test_negative))
y_test_s = np.concatenate((y_test_positive, y_test_negative))

print(x_test_s.shape)
print(y_test_s.shape)

print(x_train_s.shape)

kf = KFold(n_splits = 5, random_state=seed)
```

```
c = .05
clf = SVC(C=c,  random_state = seed)
```

```
val_accuracy_list = []
train_accuracy_list = []
count = 1
for train_index, val_index in kf.split(x_train_s):
    print("Fold : ",count)
    X_train, X_val = x_train_s[train_index], x_train_s[val_index]
    y_train, y_val = y_train_s[train_index], y_train_s[val_index]
    
    clf.fit(X_train, y_train)

    y_val_pred = clf.predict(X_val)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    val_accuracy_list.append(val_accuracy)

    y_train_pred = clf.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_accuracy_list.append(train_accuracy)

    count += 1 

print("done")
```

```
for i in range(0, len(train_accuracy_list)):
    print("train_accuracy : ", train_accuracy_list[i], "\tVal accuracy : ", val_accuracy_list[i])
```
